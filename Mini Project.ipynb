```python
#Using IQR Method to detect outliers.

#Select continuous numerical columns.
numerical_cols = [
    'Administrative_Duration',
    'Informational_Duration',
    'ProductRelated_Duration',
    'BounceRates',
    'ExitRates',
    'PageValues',
    'SpecialDay'
]

outlier_count = {}

for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    count = df[(df[col] < lower) | (df[col] > upper)].shape[0]
    outlier_count[col] = count

outlier_count
```

# **Feature Engineering:**

**Separate Features and Target**

```python
#Define target variable.
y = df['Revenue']

#Drop Revenue and unwanted columns from features.
X = df.drop(columns=[
    'Revenue',
    'Administrative_Duration',
    'Informational_Duration',
    'BounceRates'
])
```

```python
from sklearn.preprocessing import LabelEncoder

#Binary categorical columns.
binary_cols = ['Weekend']

#Multi-class categorical columns.
multi_class_cols = ['Month', 'VisitorType']

X_encoded = X.copy()

#Label Encode binary columns.
le = LabelEncoder()
for col in binary_cols:
    X_encoded[col] = le.fit_transform(X_encoded[col])

#One-Hot Encode multi-class columns.
X_encoded = pd.get_dummies(
    X_encoded,
    columns=multi_class_cols,
    drop_first=True
)

#Encode target.
y_encoded = LabelEncoder().fit_transform(y)
```

# **Feature Selection**

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import SelectKBest, f_classif

#Step 1: Temporary Encoding for Feature Selection.
X_temp = X.copy()
cat_cols = X_temp.select_dtypes(include=['object']).columns
le = LabelEncoder()
for col in cat_cols:
    X_temp[col] = le.fit_transform(X_temp[col])

#Step 2: Random Forest Feature Importance.
rf = RandomForestClassifier(n_estimators=200, random_state=42, n_jobs=-1)
rf.fit(X_temp, y)

rf_importance = pd.DataFrame({
    'Feature': X_temp.columns,
    'Importance': rf.feature_importances_
}).sort_values(by='Importance', ascending=False)

print("Random Forest Feature Importances:\n", rf_importance)

#Keep features with importance > 0.01.
rf_selected = rf_importance[rf_importance['Importance'] > 0.01]['Feature']
print("\nRF Selected Features:", list(rf_selected))

#Step 3: SelectKBest Feature Selection.
kbest = SelectKBest(score_func=f_classif, k=10)
X_kbest = kbest.fit_transform(X_temp, y)
kbest_selected = X_temp.columns[kbest.get_support()]

print("\nKBest Selected Features:", list(kbest_selected))

#Step 4: Intersection of RF & KBest.
final_features = list(set(rf_selected) & set(kbest_selected))
print("\nFinal Selected Features (Intersection):", final_features)

#Step 5: Create Final Feature Dataset.
X_final = X[final_features]
```

# 9. **Split Data into Training and Testing Sets:**

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_encoded,
    y_encoded,
    test_size=0.2,
    random_state=42,
    stratify=y_encoded
)
```

```python
#Feature Scaling – StandardScaler.

from sklearn.preprocessing import StandardScaler

#Automatically select numeric columns from the final selected features.
num_features = X_final.select_dtypes(include=['int64', 'float64']).columns.tolist()

#Initialize StandardScaler.
scaler = StandardScaler()

#Make copies of train and test sets.
X_train_scaled = X_train.copy()
X_test_scaled = X_test.copy()

#Scale only numeric columns.
X_train_scaled[num_features] = scaler.fit_transform(X_train[num_features])
X_test_scaled[num_features] = scaler.transform(X_test[num_features])


print("X_train_scaled sample:")
print(X_train_scaled.head())
print("\nX_test_scaled sample:")
print(X_test_scaled.head())
```

# 11. **Build the ML Model:**

**11A. Logistic Regression Classifier.**

```python
#Logistic Regression Model.
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

#create the model.
log_reg_model = LogisticRegression(
    max_iter=1000,
    solver='lbfgs',
    random_state=42
)

#train the model using training data.
log_reg_model.fit(X_train_scaled, y_train)

#predict on test data.
y_pred = log_reg_model.predict(X_test_scaled)

#accuracy.
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

#confusion matrix.
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", cm)

cr = classification_report(y_test, y_pred)
print("Classification Report:\n", cr)
```

```python
# ROC Curve for Logistic Regression
from sklearn.metrics import roc_curve, roc_auc_score

# probability of positive class
y_prob = log_reg_model.predict_proba(X_test_scaled)[:, 1]

#compute ROC values.
fpr, tpr, thresholds = roc_curve(y_test, y_prob)

#compute AUC score.
auc_score = roc_auc_score(y_test, y_prob)
print("ROC AUC Score:", auc_score)

#plot ROC curve.
plt.figure(figsize=(8,5))
plt.plot(fpr, tpr, label="Logistic Regression (AUC = %.3f)" % auc_score)
plt.plot([0,1], [0,1], 'k--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve - Logistic Regression")
plt.legend(loc="lower right")
plt.show()
```

**11B. SVM Classifier.**

```python
#Support Vector Machine (SVM) Model.
from sklearn.svm import SVC

#create SVM model (RBF kernel).
svm_model = SVC(
    kernel='rbf',
    C=1,
    gamma='scale',
    probability=True,
    random_state=42
)

#train the model.
svm_model.fit(X_train_scaled, y_train)

#predict class labels.
y_pred_svm = svm_model.predict(X_test_scaled)

#predict probabilities.
y_prob_svm = svm_model.predict_proba(X_test_scaled)

#accuracy.
accuracy_svm = accuracy_score(y_test, y_pred_svm)
print("Accuracy:", accuracy_svm)

#confusion matrix.
cm_svm = confusion_matrix(y_test, y_pred_svm)
print("Confusion Matrix:\n", cm_svm)

#precision, recall, f1-score.
cr_svm = classification_report(y_test, y_pred_svm)
print("Classification Report:\n", cr_svm)
```

# 13. **Hyperparameter Tuning: and  pipeline**

**13 A. Hyperparameter Tuning for Random Forest.**

```python
from sklearn.model_selection import StratifiedKFold, GridSearchCV

cv = StratifiedKFold(
    n_splits=5,
    shuffle=True,
    random_state=42
)
```

```python
from sklearn.pipeline import Pipeline

#Pipeline.
rf_pipeline = Pipeline([
    ("model", RandomForestClassifier(random_state=42))
])

#Hyperparameter grid.
param_rf = {
    "model__n_estimators": [100, 200],
    "model__max_depth": [5, 10, None],
    "model__min_samples_split": [2, 5, 10],
    "model__min_samples_leaf": [2, 5]
}

#Grid Search.
grid_rf = GridSearchCV(
    estimator=rf_pipeline,
    param_grid=param_rf,
    cv=cv,
    scoring="f1",
    n_jobs=-1,
    verbose=1,
    refit=True
)

#Fit.
grid_rf.fit(X_train, y_train)

#Best results.
print("Best Parameters:", grid_rf.best_params_)
print("Best Cross-Validation F1 Score:", grid_rf.best_score_)
```

```python
#Extract Best Tuned Model.
best_rf_model = grid_rf.best_estimator_
```
